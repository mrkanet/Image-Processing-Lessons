```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('bird.jpg')
plt.imshow(img)
plt.show()
```

```python
m,n,p=img.shape
my_histogram_R_G_B={} 
for i in range(m):
    for j in range(n):
        s=(img[i,j,1])  
        if ((1,s) in my_histogram_R_G_B.keys()):  
            my_histogram_R_G_B[(1,s)]=my_histogram_R_G_B[(1,s)]+1
        else:
            my_histogram_R_G_B[(1,s)]=1
            my_histogram_R_G_B
my_histogram_R_G_B
```

```python
histogram = {}
for i in range(m):
    for j in range(n):
        s = (img[i,j,0],img[i,j,1],img[i,j,2])
        if (s in histogram.keys()):
            histogram[s] += 1
        else:
            histogram[s] = 1
histogram
```
