```python
pixelList = []
#img2.setflags(write=1)
for i in range(1,m-1):
    for j in range(1,n-1):
        meaning = img[i-1,j-1]/9 + img[i-1,j]/9 + img[i-1,j+1]/9 + img[i,j-1]/9 + img[i,j]/9 + img[i,j+1]/9 + img[i+1,j-1] + img[i+1,j]/9 + img[i+1,j+1]/9
        #pixelList.append(meaning)
        img2[i,j] = int(meaning)
        
```
